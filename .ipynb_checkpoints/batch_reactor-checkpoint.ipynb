# Data Generation

```python
def batch_reactor_model(XA0, XB0, temp):
    """
    Solves the Williams-Otto process model to obtain the concentrations of the products

    Parameters
    ----------
    XA0: initial mass fraction of component A
    XB0: initial mass fraction of component B
    temp: temperature of the reaction

    Returns
    -------
    
    """
    model = pyo.ConcreteModel()

    # define sets
    reaction_number = [1, 2, 3]
    model.t = dae.ContinuousSet(bounds=[0, 4]) # hour

    # define the model parameters
    model.a1 = pyo.Param(within=pyo.PositiveReals, initialize=5.9755 * 10**9)
    model.a2 = pyo.Param(within=pyo.PositiveReals, initialize=2.5962 * 10**12)
    model.a3 = pyo.Param(within=pyo.PositiveReals, initialize=9.6283 * 10**15)
    model.E1 = pyo.Param(within=pyo.PositiveReals, initialize=120)
    model.E2 = pyo.Param(within=pyo.PositiveReals, initialize=150)
    model.E3 = pyo.Param(within=pyo.PositiveReals, initialize=200)
    
    # define the model variables
    model.T = pyo.Var(bounds=[5.8, 6.8],)
    model.T.fix(temp)

    model.XA = pyo.Var(model.t, bounds=[0, 1],)
    model.XB = pyo.Var(model.t, bounds=[0, 1],)
    model.XC = pyo.Var(model.t, bounds=[0, 1],)
    model.XE = pyo.Var(model.t, bounds=[0, 1],)
    model.XP = pyo.Var(model.t, bounds=[0, 1],)
    model.XG = pyo.Var(model.t, bounds=[0, 1],)

    # calculate the T2 variable
    model.T2 = pyo.Var(bounds=[0.147, 0.172])
    model.T2_eq = pyo.Constraint(expr=model.T2 == 1 / model.T)
    
    # rate constants
    model.k_reparam = pyo.Var(reaction_number, bounds=(0, None),)
    model.k = pyo.Var(reaction_number, bounds=(0, None),)

    # reaction rates
    model.r1 = pyo.Var(model.t, bounds=(0, None),)
    model.r2 = pyo.Var(model.t, bounds=(0, None),)
    model.r3 = pyo.Var(model.t, bounds=(0, None),)
    
    # calculate the reparameterized rate constants
    def k_reparam_rule(m, i):
        if i == 1:
            return m.k_reparam[i] == pyo.log(m.a1) - m.E1 * m.T2
        elif i == 2:
            return m.k_reparam[i] == pyo.log(m.a2) - m.E2 * m.T2
        else:
            return m.k_reparam[i] == pyo.log(m.a3) - m.E3 * m.T2
            
    model.k_reparam_eq = pyo.Constraint(
        reaction_number, rule=k_reparam_rule
    )
    
    # calculate the original rate constants
    def k_rule(m, i):
        return m.k[i] == pyo.exp(m.k_reparam[i])
    
    model.k_eq = pyo.Constraint(
        reaction_number, rule=k_rule
    )
    
    # define the rates of reactions
    def reaction_rate1_rule(m, t):
        return m.r1[t] == m.k[1] * m.XA[t] * m.XB[t]

    def reaction_rate2_rule(m, t):
        return m.r2[t] == m.k[2] * m.XB[t] * m.XC[t]
    
    def reaction_rate3_rule(m, t):
        return m.r3[t] == m.k[3] * m.XC[t] * m.XP[t]
    
    model.reaction_rate1 = pyo.Constraint(
        model.t, rule=reaction_rate1_rule
    )
    model.reaction_rate2 = pyo.Constraint(
        model.t, rule=reaction_rate2_rule
    )
    model.reaction_rate3 = pyo.Constraint(
        model.t, rule=reaction_rate3_rule
    )

    # add the differential equations for XA, XB, XC, XE, XP, and XG
    model.dXA = dae.DerivativeVar(model.XA, wrt=model.t)
    model.dXB = dae.DerivativeVar(model.XB, wrt=model.t)
    model.dXC = dae.DerivativeVar(model.XC, wrt=model.t)
    model.dXE = dae.DerivativeVar(model.XE, wrt=model.t)
    model.dXP = dae.DerivativeVar(model.XP, wrt=model.t)
    model.dXG = dae.DerivativeVar(model.XG, wrt=model.t)

    @model.Constraint(model.t)
    def xa_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXA[t] == - m.k[1] * m.XA[t] * m.XB[t]
    
    @model.Constraint(model.t)
    def xb_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXB[t] == - (m.k[1] * m.XA[t] * m.XB[t] + m.k[2] * m.XB[t] * m.XC[t])

    @model.Constraint(model.t)
    def xc_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXC[t] == 2 * m.k[1] * m.XA[t] * m.XB[t] - 2 * m.k[2] * m.XB[t] * m.XC[t] - m.k[3] * m.XC[t] * m.XP[t]

    @model.Constraint(model.t)
    def xe_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXE[t] == 2 * m.k[2] * m.XB[t] * m.XC[t]
    
    @model.Constraint(model.t)
    def xg_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXG[t] == 1.5 * m.k[3] * m.XC[t] * m.XP[t]

    @model.Constraint(model.t)
    def xp_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXP[t] == m.k[2] * m.XB[t] * m.XC[t] - 0.5 * m.k[3] * m.XC[t] * m.XP[t]

    # # mass fraction constraint
    # @model.Constraint(model.t)
    # def sum_mass_fraction(m, t):
    #     if t == m.t.first():
    #         return pyo.Constraint.Skip
    #     return m.XA[t] + m.XB[t] + m.XC[t] + m.XE[t] + m.XG[t] + m.XP[t] == 1

    # initial conditions
    model.XA[0].fix(XA0)
    model.XB[0].fix(XB0)
    model.XC[0].fix(0)
    model.XE[0].fix(0)
    model.XP[0].fix(0)
    model.XG[0].fix(0)
    # t0 = model.t.first()
    # model.XA_init = pyo.Constraint(expr=model.XA[t0] == XA0)
    # model.XB_init = pyo.Constraint(expr=model.XB[t0] == XB0)
    # model.XC_init = pyo.Constraint(expr=model.XC[t0] == 0.0)
    # model.XE_init = pyo.Constraint(expr=model.XE[t0] == 0.0)
    # model.XP_init = pyo.Constraint(expr=model.XP[t0] == 0.0)
    # model.XG_init = pyo.Constraint(expr=model.XG[t0] == 0.0)

    # disc = pyo.TransformationFactory("dae.collocation")
    # disc.apply_to(model, nfe=30, ncp=3)
    disc = pyo.TransformationFactory("dae.finite_difference")
    disc.apply_to(model, nfe=80, scheme="BACKWARD")
    
    # solve the model
    solver = pyo.SolverFactory('ipopt')
    
    # get the solved results
    results = solver.solve(model, tee=True)

    time = [t for t in model.t]
    print("Time",time)
    XA_values = [pyo.value(model.XA[t]) for t in model.t]
    XB_values = [pyo.value(model.XB[t]) for t in model.t] 
    XP_values = [pyo.value(model.XP[t]) for t in model.t] 
    XE_values = [pyo.value(model.XE[t]) for t in model.t] 
    XG_values = [pyo.value(model.XG[t]) for t in model.t] 
    
    # Plot the results
    plt.figure(figsize=(8, 6))
    plt.plot(time, XA_values, label='$\mathrm{X_{A}}$')
    plt.plot(time, XB_values, label='$\mathrm{X_{B}}$')
    plt.plot(time, XP_values, label='$\mathrm{X_{P}}$')
    plt.plot(time, XE_values, label='$\mathrm{X_{E}}$')
    plt.plot(time, XG_values, label='$\mathrm{X_{G}}$')
    plt.xlabel('Time (hr)', fontsize=16,fontweight='bold')
    plt.ylabel('Mass Fraction', fontsize=16,fontweight='bold')
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(prop={"weight":"bold", "size": 11}, bbox_to_anchor=(1, 0.5))
    plt.savefig("batch_reactor.png")
    plt.show()

    # data_time_points = [0, 0.1, 0.2, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4,]
    data_time_points = time
    measured_variables = []
    for time_point in data_time_points:
        measured_variables.append([pyo.value(model.XA[time_point]), pyo.value(model.XB[time_point]), pyo.value(model.XP[time_point]), 
                                   pyo.value(model.XE[time_point]), pyo.value(model.XG[time_point])])

    print("Measured variables:",measured_variables)
    return data_time_points, measured_variables
```

```python
time_points, measured_var = batch_reactor_model(0.4, 0.6, 6)
```

```python
def batch_reactor_model_exp(XA0, XB0, temp):
    """
    Solves the Williams-Otto process model to obtain the concentrations of the products

    Parameters
    ----------
    XA0: initial mass fraction of component A
    XB0: initial mass fraction of component B
    temp: temperature of the reaction

    Returns
    -------
    
    """
    model = pyo.ConcreteModel()

    # define sets
    reaction_number = [1, 2, 3]
    model.t = dae.ContinuousSet(bounds=[0, 4]) # hour

    # define the model parameters
    model.a1 = pyo.Var(within=pyo.NonNegativeReals, initialize=1e6)
    model.a2 = pyo.Var(within=pyo.NonNegativeReals, initialize=1e6)
    model.a3 = pyo.Var(within=pyo.NonNegativeReals, initialize=1e6)
    # model.log_a1 = pyo.Var(within=pyo.PositiveReals, initialize=10)
    # model.log_a2 = pyo.Var(within=pyo.PositiveReals, initialize=10)
    # model.log_a3 = pyo.Var(within=pyo.PositiveReals, initialize=10)
    
    model.E1 = pyo.Var(within=pyo.NonNegativeReals, initialize=100)
    model.E2 = pyo.Var(within=pyo.NonNegativeReals, initialize=100)
    model.E3 = pyo.Var(within=pyo.NonNegativeReals, initialize=100)
    
    # define the model variables
    model.T = pyo.Var(bounds=[5.8, 6.8],)
    model.T.fix(temp)
    
    model.XA = pyo.Var(model.t, bounds=[0, 1], initialize=XA0)
    model.XB = pyo.Var(model.t, bounds=[0, 1], initialize=XB0)
    model.XC = pyo.Var(model.t, bounds=[0, 1],)
    model.XE = pyo.Var(model.t, bounds=[0, 1],)
    model.XP = pyo.Var(model.t, bounds=[0, 1],)
    model.XG = pyo.Var(model.t, bounds=[0, 1],)

    # calculate the T2 variable
    model.T2 = pyo.Var(bounds=[0.147, 0.172],)
    model.T2_eq = pyo.Constraint(expr=model.T2 == 1 / model.T)
    
    # rate constants
    model.k_reparam = pyo.Var(reaction_number, bounds=(0, None),)
    model.k = pyo.Var(reaction_number, bounds=(0, None),)

    # reaction rates
    model.r1 = pyo.Var(model.t, bounds=(0, None),)
    model.r2 = pyo.Var(model.t, bounds=(0, None),)
    model.r3 = pyo.Var(model.t, bounds=(0, None),)
    
    # calculate the reparameterized rate constants
    def k_reparam_rule(m, i):
        if i == 1:
            return m.k_reparam[i] == pyo.log(m.a1) - m.E1 * m.T2
        elif i == 2:
            return m.k_reparam[i] == pyo.log(m.a2) - m.E2 * m.T2
        else:
            return m.k_reparam[i] == pyo.log(m.a3) - m.E3 * m.T2
            
    model.k_reparam_eq = pyo.Constraint(
        reaction_number, rule=k_reparam_rule
    )
    
    # calculate the original rate constants
    def k_rule(m, i):
        return m.k[i] == pyo.exp(m.k_reparam[i])
    
    model.k_eq = pyo.Constraint(
        reaction_number, rule=k_rule
    )
    
    # define the rates of reactions
    def reaction_rate1_rule(m, t):
        return m.r1[t] == m.k[1] * m.XA[t] * m.XB[t]

    def reaction_rate2_rule(m, t):
        return m.r2[t] == m.k[2] * m.XB[t] * m.XC[t]
    
    def reaction_rate3_rule(m, t):
        return m.r3[t] == m.k[3] * m.XC[t] * m.XP[t]
    
    model.reaction_rate1 = pyo.Constraint(
        model.t, rule=reaction_rate1_rule
    )
    model.reaction_rate2 = pyo.Constraint(
        model.t, rule=reaction_rate2_rule
    )
    model.reaction_rate3 = pyo.Constraint(
        model.t, rule=reaction_rate3_rule
    )

    # add the differential equations for XA, XB, XC, XE, XP, and XG
    model.dXA = dae.DerivativeVar(model.XA, wrt=model.t)
    model.dXB = dae.DerivativeVar(model.XB, wrt=model.t)
    model.dXC = dae.DerivativeVar(model.XC, wrt=model.t)
    model.dXE = dae.DerivativeVar(model.XE, wrt=model.t)
    model.dXP = dae.DerivativeVar(model.XP, wrt=model.t)
    model.dXG = dae.DerivativeVar(model.XG, wrt=model.t)

    @model.Constraint(model.t)
    def xa_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXA[t] == - m.k[1] * m.XA[t] * m.XB[t]
    
    @model.Constraint(model.t)
    def xb_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXB[t] == - (m.k[1] * m.XA[t] * m.XB[t] + m.k[2] * m.XB[t] * m.XC[t])

    @model.Constraint(model.t)
    def xc_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXC[t] == 2 * m.k[1] * m.XA[t] * m.XB[t] - 2 * m.k[2] * m.XB[t] * m.XC[t] - m.k[3] * m.XC[t] * m.XP[t]

    @model.Constraint(model.t)
    def xe_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXE[t] == 2 * m.k[2] * m.XB[t] * m.XC[t]
    
    @model.Constraint(model.t)
    def xg_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXG[t] == 1.5 * m.k[3] * m.XC[t] * m.XP[t]

    @model.Constraint(model.t)
    def xp_rate_ode(m, t):
        if t == m.t.first():
            return pyo.Constraint.Skip
        return m.dXP[t] == m.k[2] * m.XB[t] * m.XC[t] - 0.5 * m.k[3] * m.XC[t] * m.XP[t]

    # initial conditions
    t0 = model.t.first()
    model.XA_init = pyo.Constraint(expr=model.XA[t0] == XA0)
    model.XB_init = pyo.Constraint(expr=model.XB[t0] == XB0)
    model.XC_init = pyo.Constraint(expr=model.XC[t0] == 0.0)
    model.XE_init = pyo.Constraint(expr=model.XE[t0] == 0.0)
    model.XP_init = pyo.Constraint(expr=model.XP[t0] == 0.0)
    model.XG_init = pyo.Constraint(expr=model.XG[t0] == 0.0)

    disc = pyo.TransformationFactory("dae.finite_difference")
    disc.apply_to(model, nfe=80, scheme="BACKWARD")
    
    # solve the model
    solver = pyo.SolverFactory('ipopt')
    
    # get the solved results
    results = solver.solve(model, tee=True)
    
    return model
```

```python
class BatchReactorExperiment(Experiment):

    def __init__(self, data, XA0, XB0, temp):

        self.data = data
        self.XA0 = XA0
        self.XB0 = XB0
        self.temp = temp
        self.model = None

    def get_labeled_model(self):
        self.create_model()
        self.label_model()

        return self.model

    def create_model(self):
        self.model = batch_reactor_model_exp(self.XA0, self.XB0, self.temp)

        return self.model

    def label_model(self):

        m = self.model

        meas_time_points = self.data["Time"]

        # label the measured variables
        m.experiment_outputs = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        m.experiment_outputs.update(
            (m.XA[t], self.data["XA"][ind]) for ind, t in enumerate(meas_time_points)
        )
        m.experiment_outputs.update(
            (m.XB[t], self.data["XB"][ind]) for ind, t in enumerate(meas_time_points)
        )
        m.experiment_outputs.update(
            (m.XP[t], self.data["XP"][ind]) for ind, t in enumerate(meas_time_points)
        )
        m.experiment_outputs.update(
            (m.XE[t], self.data["XE"][ind]) for ind, t in enumerate(meas_time_points)
        )
        m.experiment_outputs.update(
            (m.XG[t], self.data["XG"][ind]) for ind, t in enumerate(meas_time_points)
        )

        # label the unknown parameters
        m.unknown_parameters = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        m.unknown_parameters.update(
            (k, pyo.ComponentUID(k)) for k in [m.a1, m.a2, m.a3, m.E1, m.E2, m.E3]
        )

        # add the measurement error
        m.measurement_error = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        m.measurement_error.update(
            (m.XA[t], mass_frac_std) for t in meas_time_points
        )
        m.measurement_error.update(
            (m.XB[t], mass_frac_std) for t in meas_time_points
        )
        m.measurement_error.update(
            (m.XP[t], mass_frac_std) for t in meas_time_points
        )
        m.measurement_error.update(
            (m.XE[t], mass_frac_std) for t in meas_time_points
        )
        m.measurement_error.update(
            (m.XG[t], mass_frac_std) for t in meas_time_points
        )

        return m
```

```python
# create the list of the experiment class
exp_list = [BatchReactorExperiment(data, 0.4, 0.6, 6)]

# Create the Estimator object
pest = parmest.Estimator(exp_list, obj_function="SSE_weighted", tee=True)  

# Estimate the parameters
obj, theta = pest.theta_est()
print("The estimated parameters are:", theta)
print("The objective value is:", obj)

# calculate the covariance matrix
cov = pest.cov_est(method="finite_difference")

print("The covariance matrix is :", cov)
```
